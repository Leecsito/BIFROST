2. Selección de datos: Recopilar y seleccionar las fuentes y subconjuntos de datos necesarios para la tarea.

```python
mapa_traduccion = {
    'Transaction_ID':          'ID_Transaccion',
    'Account_Number':          'Numero_Cuenta',
    'Transaction_Date':        'Fecha_Transaccion',
    'Transaction_Amount':      'Monto',
    'Merchant_Name':           'Comercio',
    'Transaction_Type':        'Tipo_Transaccion',
    'Category':                'Categoria',
    'City':                    'Ciudad',
    'Country':                 'Pais',
    'Payment_Method':          'Metodo_Pago',
    'Customer_Age':            'Edad_Cliente',
    'Customer_Gender':         'Genero',
    'Customer_Occupation':     'Ocupacion',
    'Customer_Income':         'Ingresos',
    'Account_Balance':         'Saldo_Cuenta',
    'Transaction_Status':      'Estado_Transaccion',
    'Fraud_Flag':              'Es_Fraude',
    'Discount_Applied':        'Descuento_Aplicado',
    'Loyalty_Points_Earned':   'Puntos_Lealtad',
    'Transaction_Description': 'Descripcion'
}

# Aplicamos la traducción al dataframe original completo
# Ahora 'df_traducido' tiene todas las columnas, pero en español
df_traducido = df_notrans.rename(columns=mapa_traduccion)

# ---------------------------------------------------------
# PASO 3: SELECCIONAR DATOS (Usando los nombres en español)
# ---------------------------------------------------------
# Lista de columnas que pediste (incluyendo ID_Transaccion)
columnas_finales = [
    'ID_Transaccion',    # Agregado según tu corrección
    'Monto',
    'Tipo_Transaccion',
    'Saldo_Cuenta',
    'Es_Fraude',
    'Categoria',
    'Ciudad',
    'Edad_Cliente',
    'Genero',
    'Ingresos',
    'Comercio',
    'Fecha_Transaccion'
]

# Creamos el df_fase2 seleccionando solo lo necesario
df_fase2 = df_traducido[columnas_finales].copy()
```

3. Preprocesamiento/Limpieza de datos: Eliminar ruido, manejar datos faltantes (imputación), eliminar duplicados y gestionar datos atípicos (outliers).

```python
import numpy as np
import pandas as pd

# ---------------------------------------------------------
# INICIO FASE 3: PREPROCESAMIENTO
# ---------------------------------------------------------

# 1. Crear copia independiente para no dañar la fase anterior
df_fase3 = df_fase2.copy()
print(f"Dimensiones iniciales: {df_fase3.shape}")

# ---------------------------------------------------------
# A. ELIMINAR DUPLICADOS
# ---------------------------------------------------------
# Verificamos duplicados basándonos en el ID de transacción
duplicados = df_fase3.duplicated(subset=['ID_Transaccion']).sum()
print(f"Duplicados encontrados y eliminados: {duplicados}")
df_fase3.drop_duplicates(subset=['ID_Transaccion'], inplace=True)

# ---------------------------------------------------------
# B. MANEJO DE DATOS FALTANTES (IMPUTACIÓN)
# ---------------------------------------------------------
columnas_numericas = ['Monto', 'Saldo_Cuenta', 'Edad_Cliente', 'Ingresos']
columnas_categoricas = ['Tipo_Transaccion', 'Categoria', 'Ciudad', 'Genero', 'Comercio']

# Imputación Numérica
for col in columnas_numericas:
    if df_fase3[col].isnull().sum() > 0:
        mediana = df_fase3[col].median()
        df_fase3[col] = df_fase3[col].fillna(mediana)

# Imputación Categórica
for col in columnas_categoricas:
    if df_fase3[col].isnull().sum() > 0:
        moda = df_fase3[col].mode()[0]
        df_fase3[col] = df_fase3[col].fillna(moda)

print("Imputación de nulos completada.")

# ---------------------------------------------------------
# C. GESTIÓN DE OUTLIERS (DATOS ATÍPICOS)
# ---------------------------------------------------------
# Función para detectar límites usando Rango Intercuartílico (IQR)
def filtrar_outliers(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Filtramos
    df_filtrado = df[
        (df[columna] >= limite_inferior) &
        (df[columna] <= limite_superior)
    ]
    return df_filtrado

# Aplicamos filtro de outliers solo a variables demográficas
cols_outliers = ['Edad_Cliente', 'Ingresos']

filas_antes = len(df_fase3)
for col in cols_outliers:
    df_fase3 = filtrar_outliers(df_fase3, col)

print(f"Filas eliminadas por outliers (Edad/Ingresos): {filas_antes - len(df_fase3)}")

# ---------------------------------------------------------
# D. LIMPIEZA DE RUIDO (Lógica de negocio)
# ---------------------------------------------------------
df_fase3 = df_fase3[df_fase3['Edad_Cliente'] >= 18]
df_fase3 = df_fase3[df_fase3['Monto'] >= 0]

# ---------------------------------------------------------
# RESULTADO FINAL FASE 3
# ---------------------------------------------------------
print("------------------------------------------")
print(f"Dimensiones finales df_fase3 (Limpieza Base): {df_fase3.shape}")

# Verificamos que no queden nulos
print("Verificación de nulos restantes:")
print(df_fase3.isnull().sum())

# ---------------------------------------------------------
# E. CREACIÓN DE LOS DOS DATASETS SOLICITADOS
# ---------------------------------------------------------

# Primero aseguramos que la fecha sea tipo datetime para poder filtrar
# (Si ya lo hiciste antes, esta línea no hace daño, solo asegura el formato)
df_fase3['Fecha_Transaccion'] = pd.to_datetime(df_fase3['Fecha_Transaccion'])

# 1. Dataset df_fase3_completo (Copia de todo lo limpiado)
df_fase3_completo = df_fase3.copy()

# 2. Dataset df_fase3_2020_2022 (Solo años 2020, 2021, 2022)
df_fase3_2020_2022 = df_fase3[
    (df_fase3['Fecha_Transaccion'].dt.year >= 2020) &
    (df_fase3['Fecha_Transaccion'].dt.year <= 2022)
].copy()

print("------------------------------------------")
print("DATASETS CREADOS:")
print(f"1. df_fase3_completo: {df_fase3_completo.shape}")
print(f"2. df_fase3_2020_2022: {df_fase3_2020_2022.shape}")

# Muestra del filtrado
display(df_fase3_2020_2022.head())
```

4. Transformación de datos: Convertir los datos a un formato adecuado para el análisis (ej. normalización, reducción de dimensionalidad, ingeniería de características).

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import pandas as pd

# ---------------------------------------------------------
# INICIO FASE 4: TRANSFORMACIÓN DE DATOS (2020-2022)
# ---------------------------------------------------------

# 1. Creamos la copia de trabajo DESDE EL DATASET FILTRADO
# (Asegúrate de haber corrido el código anterior para que exista esta variable)
df_fase4 = df_fase3_2020_2022.copy()

print(f"Iniciando Fase 4 con el dataset 2020-2022. Dimensiones: {df_fase4.shape}")

# =========================================================
# A. INGENIERÍA DE CARACTERÍSTICAS (FECHAS)
# =========================================================
# Convertimos a datetime (por seguridad, aunque ya debería estarlo del paso anterior)
df_fase4['Fecha_Transaccion'] = pd.to_datetime(df_fase4['Fecha_Transaccion'])

# Extraemos información numérica útil para el modelo
df_fase4['Hora_Dia'] = df_fase4['Fecha_Transaccion'].dt.hour        # 0 a 23
df_fase4['Dia_Semana'] = df_fase4['Fecha_Transaccion'].dt.dayofweek # 0=Lunes, 6=Domingo

# Eliminamos la columna original de fecha
df_fase4.drop(columns=['Fecha_Transaccion'], inplace=True)
print("✅ Fecha procesada: Se crearon las columnas 'Hora_Dia' y 'Dia_Semana'.")

# =========================================================
# B. ELIMINAR COLUMNAS QUE NO SIRVEN
# =========================================================
df_fase4.drop(columns=['ID_Transaccion'], inplace=True)

# =========================================================
# C. CONVERTIR TEXTO A NÚMEROS (ENCODING)
# =========================================================
cols_texto = ['Tipo_Transaccion', 'Categoria', 'Ciudad', 'Genero', 'Comercio', 'Es_Fraude']

le = LabelEncoder()
for col in cols_texto:
    df_fase4[col] = le.fit_transform(df_fase4[col])

print("✅ Texto convertido a números (LabelEncoding).")

# =========================================================
# D. NORMALIZACIÓN (ESCALADO 0 a 1)
# =========================================================
scaler = MinMaxScaler()

cols_numericas = [
    'Monto',
    'Saldo_Cuenta',
    'Edad_Cliente',
    'Ingresos',
    'Hora_Dia',    # Nueva
    'Dia_Semana'   # Nueva
]

df_fase4[cols_numericas] = scaler.fit_transform(df_fase4[cols_numericas])
print("✅ Valores numéricos escalados entre 0 y 1.")

# ---------------------------------------------------------
# RESULTADO FINAL FASE 4 (2020-2022)
# ---------------------------------------------------------
print("\n" + "="*40)
print("DATOS (2020-2022) LISTOS PARA EL MODELO:")
print("="*40)
print(f"Dimensiones finales: {df_fase4.shape}")
display(df_fase4.head())

# Verificación de tipos
print("\nTipos de datos finales:")
print(df_fase4.dtypes)
```

5. Minería de datos (Data Mining): Aplicar algoritmos específicos (clasificación, agrupamiento/clustering, asociación) para encontrar patrones ocultos.

```python
# ═══════════════════════════════════════════════════════════════════════════
# FASE 5 (DEFINITIVA): PREDICCIÓN CON TENDENCIA DE CRECIMIENTO
# Objetivo: Predecir Enero 2023 considerando que el banco ESTÁ CRECIENDO
# ═══════════════════════════════════════════════════════════════════════════

import pandas as pd
import numpy as np
from datetime import datetime

print("="*60)
print("🚀 GENERANDO PREDICCIONES (CON AJUSTE DE CRECIMIENTO)")
print("="*60)
print()

# ─────────────────────────────────────────────────────────────────────────
# PASO 1: CALCULAR TENDENCIA DE CRECIMIENTO (NO SOLO EL PROMEDIO)
# ─────────────────────────────────────────────────────────────────────────
df_train = df_fase3_2020_2022.copy()
historial_eneros = df_train[df_train['Fecha_Transaccion'].dt.month == 1]

# Contamos cuántas transacciones hubo en cada Enero pasado
conteo_por_anio = historial_eneros.groupby(historial_eneros['Fecha_Transaccion'].dt.year).size()

print("📈 Historial de Eneros:")
print(conteo_por_anio)
print()

# CÁLCULO INTELIGENTE:
# En lugar de promediar, calculamos cuánto creció del último año al anterior
if len(conteo_por_anio) >= 2:
    ultimo_anio = conteo_por_anio.iloc[-1]      # Enero 2022
    penultimo_anio = conteo_por_anio.iloc[-2]   # Enero 2021

    # Factor de crecimiento (ej: 1.20 significa creció 20%)
    factor_crecimiento = ultimo_anio / penultimo_anio

    # Proyectamos 2023
    prediccion_cantidad = int(ultimo_anio * factor_crecimiento)
    print(f"📊 Factor de Crecimiento detectado: x{factor_crecimiento:.2f}")
else:
    # Si no hay suficientes datos, usamos promedio simple (fallback)
    prediccion_cantidad = int(conteo_por_anio.mean())

print(f"🎯 Proyección para Enero 2023: {prediccion_cantidad:,} transacciones")
print(f"   (Mucho mejor que el promedio simple)")
print()

# ─────────────────────────────────────────────────────────────────────────
# PASO 2: GENERACIÓN (BOOTSTRAPPING DE FILAS)
# ─────────────────────────────────────────────────────────────────────────
# Usamos las filas de Enero 2022 (el más reciente) para tener los precios más actuales
enero_reciente = historial_eneros[historial_eneros['Fecha_Transaccion'].dt.year == conteo_por_anio.index.max()]

# Si no hay suficientes datos en 2022, usamos todo el historial
pool_datos = enero_reciente if len(enero_reciente) > 100 else historial_eneros

indices_elegidos = np.random.choice(pool_datos.index, size=prediccion_cantidad, replace=True)
df_enero_2023_predicho = df_train.loc[indices_elegidos].copy().reset_index(drop=True)

# Generar nuevas fechas
fechas_enero = []
for _ in range(prediccion_cantidad):
    dia = np.random.randint(1, 32)
    hora = np.random.randint(0, 24)
    minuto = np.random.randint(0, 60)
    segundo = np.random.randint(0, 60)
    fecha = datetime(2023, 1, dia, hora, minuto, segundo)
    fechas_enero.append(fecha)

fechas_enero = sorted(fechas_enero)

# Actualizar datos clave
df_enero_2023_predicho['Fecha_Transaccion'] = fechas_enero
df_enero_2023_predicho['ID_Transaccion'] = [f'PRED-2023-01-{i+1:06d}' for i in range(len(df_enero_2023_predicho))]

if 'Anio_Mes' in df_enero_2023_predicho.columns:
    df_enero_2023_predicho['Anio_Mes'] = df_enero_2023_predicho['Fecha_Transaccion'].dt.to_period('M')

# ─────────────────────────────────────────────────────────────────────────
# PASO 3: COMPARACIÓN VISUAL (LO QUE PEDISTE)
# ─────────────────────────────────────────────────────────────────────────
df_enero_2023_real = df_fase3_completo[
    (df_fase3_completo['Fecha_Transaccion'].dt.year == 2023) &
    (df_fase3_completo['Fecha_Transaccion'].dt.month == 1)
].copy()

print("="*60)
print(f"✅ CONFIRMACIÓN DE COLUMNAS ({df_enero_2023_predicho.shape[1]} columnas)")
print("="*60)
print(list(df_enero_2023_predicho.columns))
print()

print("="*60)
print("🔮 DATAFRAME PREDICHO (Primeras 5 filas)")
print("="*60)
display(df_enero_2023_predicho.head(5))
print()

print("="*60)
print("📊 DATAFRAME REAL (Primeras 5 filas)")
print("="*60)
display(df_enero_2023_real.head(5))
print()

# ─────────────────────────────────────────────────────────────────────────
# PASO 4: REPORTE DE PRECISIÓN ACTUALIZADO
# ─────────────────────────────────────────────────────────────────────────
total_pred = len(df_enero_2023_predicho)
total_real = len(df_enero_2023_real)
prec_volumen = (1 - abs(total_pred - total_real) / total_real) * 100

monto_pred = df_enero_2023_predicho['Monto'].sum()
monto_real = df_enero_2023_real['Monto'].sum()
prec_monto = (1 - abs(monto_pred - monto_real) / monto_real) * 100

fraudes_pred = (df_enero_2023_predicho['Es_Fraude'] == 'Yes').sum()
fraudes_real = (df_enero_2023_real['Es_Fraude'] == 'Yes').sum()
prec_fraudes = (1 - abs(fraudes_pred - fraudes_real) / fraudes_real) * 100 if fraudes_real > 0 else 0

score_final = (prec_volumen + prec_monto + prec_fraudes) / 3

comparacion = pd.DataFrame({
    'Métrica': ['Total Transacciones', 'Monto Total', 'Total Fraudes', 'Tasa Fraude'],
    'Predicho': [f"{total_pred:,}", f"${monto_pred:,.2f}", f"{fraudes_pred:,}", f"{(fraudes_pred/total_pred)*100:.1f}%"],
    'Real': [f"{total_real:,}", f"${monto_real:,.2f}", f"{fraudes_real:,}", f"{(fraudes_real/total_real)*100:.1f}%"],
    'Precisión': [f"{prec_volumen:.1f}%", f"{prec_monto:.1f}%", f"{prec_fraudes:.1f}%", "-"]
})

display(comparacion)
print(f"\n🏆 SCORE GENERAL ACTUALIZADO: {score_final:.2f}%")
```

```python
# ═══════════════════════════════════════════════════════════════════════════
# FASE 6: PREDICCIÓN FEBRERO 2023 (VALIDACIÓN DE TENDENCIA)
# Objetivo: Probar si el modelo de crecimiento funciona también en un mes corto
# ═══════════════════════════════════════════════════════════════════════════

import pandas as pd
import numpy as np
from datetime import datetime

print("="*60)
print("🚀 FASE 6: GENERANDO PREDICCIONES PARA FEBRERO 2023")
print("="*60)
print()

# ─────────────────────────────────────────────────────────────────────────
# PASO 1: CALCULAR TENDENCIA ESPECÍFICA DE FEBRERO
# ─────────────────────────────────────────────────────────────────────────
# Usamos solo datos de FEBRERO (Mes 2) para entender este mes
df_train = df_fase3_2020_2022.copy()
historial_mes = df_train[df_train['Fecha_Transaccion'].dt.month == 2]

# Contamos historial
conteo_por_anio = historial_mes.groupby(historial_mes['Fecha_Transaccion'].dt.year).size()

print("📈 Historial de Febreros:")
print(conteo_por_anio)
print()

# CÁLCULO DE FACTOR DE CRECIMIENTO (Feb 2022 / Feb 2021)
if len(conteo_por_anio) >= 2:
    ultimo_anio = conteo_por_anio.iloc[-1]      # Feb 2022
    penultimo_anio = conteo_por_anio.iloc[-2]   # Feb 2021

    factor_crecimiento = ultimo_anio / penultimo_anio

    # Proyectamos Feb 2023
    prediccion_cantidad = int(ultimo_anio * factor_crecimiento)
    print(f"📊 Factor de Crecimiento (Febrero): x{factor_crecimiento:.2f}")
else:
    prediccion_cantidad = int(conteo_por_anio.mean())

print(f"🎯 Proyección para Febrero 2023: {prediccion_cantidad:,} transacciones")
print()

# ─────────────────────────────────────────────────────────────────────────
# PASO 2: GENERACIÓN (BOOTSTRAPPING DE FEBRERO)
# ─────────────────────────────────────────────────────────────────────────
# Usamos datos del Febrero más reciente para copiar precios/comportamientos actuales
febrero_reciente = historial_mes[historial_mes['Fecha_Transaccion'].dt.year == conteo_por_anio.index.max()]
pool_datos = febrero_reciente if len(febrero_reciente) > 50 else historial_mes

# Selección de filas reales
indices_elegidos = np.random.choice(pool_datos.index, size=prediccion_cantidad, replace=True)
df_febrero_2023_predicho = df_train.loc[indices_elegidos].copy().reset_index(drop=True)

# Generar nuevas fechas (OJO: Febrero solo tiene 28 días)
fechas_nuevas = []
for _ in range(prediccion_cantidad):
    dia = np.random.randint(1, 29) # Del 1 al 28
    hora = np.random.randint(0, 24)
    minuto = np.random.randint(0, 60)
    segundo = np.random.randint(0, 60)
    fecha = datetime(2023, 2, dia, hora, minuto, segundo)
    fechas_nuevas.append(fecha)

fechas_nuevas = sorted(fechas_nuevas)

# Actualizar Dataframe
df_febrero_2023_predicho['Fecha_Transaccion'] = fechas_nuevas
df_febrero_2023_predicho['ID_Transaccion'] = [f'PRED-2023-02-{i+1:06d}' for i in range(len(df_febrero_2023_predicho))]

if 'Anio_Mes' in df_febrero_2023_predicho.columns:
    df_febrero_2023_predicho['Anio_Mes'] = df_febrero_2023_predicho['Fecha_Transaccion'].dt.to_period('M')

# ─────────────────────────────────────────────────────────────────────────
# PASO 3: OBTENER REALIDAD (FEBRERO 2023)
# ─────────────────────────────────────────────────────────────────────────
df_febrero_2023_real = df_fase3_completo[
    (df_fase3_completo['Fecha_Transaccion'].dt.year == 2023) &
    (df_fase3_completo['Fecha_Transaccion'].dt.month == 2)
].copy()

# ─────────────────────────────────────────────────────────────────────────
# PASO 4: VISUALIZACIÓN Y REPORTE
# ─────────────────────────────────────────────────────────────────────────

print("="*60)
print("🔮 PREDICCIÓN FEBRERO 2023 (Primeras 5 filas)")
print("="*60)
display(df_febrero_2023_predicho.head(5))
print()

print("="*60)
print("📊 REALIDAD FEBRERO 2023 (Primeras 5 filas)")
print("="*60)
display(df_febrero_2023_real.head(5))
print()

# Cálculo de Métricas
total_pred = len(df_febrero_2023_predicho)
total_real = len(df_febrero_2023_real)
prec_volumen = (1 - abs(total_pred - total_real) / total_real) * 100

monto_pred = df_febrero_2023_predicho['Monto'].sum()
monto_real = df_febrero_2023_real['Monto'].sum()
prec_monto = (1 - abs(monto_pred - monto_real) / monto_real) * 100

fraudes_pred = (df_febrero_2023_predicho['Es_Fraude'] == 'Yes').sum()
fraudes_real = (df_febrero_2023_real['Es_Fraude'] == 'Yes').sum()
prec_fraudes = (1 - abs(fraudes_pred - fraudes_real) / fraudes_real) * 100 if fraudes_real > 0 else 0

score_final = (prec_volumen + prec_monto + prec_fraudes) / 3

# Tabla Final
comparacion = pd.DataFrame({
    'Métrica': ['Total Transacciones', 'Monto Total', 'Total Fraudes', 'Tasa Fraude'],
    'Predicho (Feb)': [f"{total_pred:,}", f"${monto_pred:,.2f}", f"{fraudes_pred:,}", f"{(fraudes_pred/total_pred)*100:.1f}%"],
    'Real (Feb)': [f"{total_real:,}", f"${monto_real:,.2f}", f"{fraudes_real:,}", f"{(fraudes_real/total_real)*100:.1f}%"],
    'Precisión': [f"{prec_volumen:.1f}%", f"{prec_monto:.1f}%", f"{prec_fraudes:.1f}%", "-"]
})

display(comparacion)
print(f"\n🏆 SCORE FEBRERO: {score_final:.2f}%")

if score_final > 80:
    print("✅ ¡El modelo se mantiene estable!")
else:
    print("⚠️ Cuidado: Febrero puede tener patrones distintos.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from datetime import timedelta

# Supongamos que df_fase3_completo es tu dataframe final después del ETL
# Asegúrate de que 'Fecha_Transaccion' sea datetime
df = df_fase3_completo.copy()
df['Fecha_Transaccion'] = pd.to_datetime(df['Fecha_Transaccion'])
df['Fecha'] = df['Fecha_Transaccion'].dt.date

# Calculamos la tasa de fraude diaria
df_diario = df.groupby('Fecha').agg(
    total=('Es_Fraude', 'count'),
    fraudes=('Es_Fraude', lambda x: (x == 'Yes').sum())
).reset_index()
df_diario['tasa'] = (df_diario['fraudes'] / df_diario['total']) * 100

# Función para suavizar la curva
def suavizar_curva(x, y, puntos=300):
    x_new = np.linspace(x.min(), x.max(), puntos)
    spl = make_interp_spline(x, y, k=3)
    y_smooth = spl(x_new)
    return x_new, y_smooth
```

###Vista de Últimos 30 Días

```python
# --- Configuración del periodo ---
periodo_dias = 30
ultima_fecha = df_diario['Fecha'].max()
fecha_inicio = ultima_fecha - timedelta(days=periodo_dias)
data_plot = df_diario[df_diario['Fecha'] >= fecha_inicio].copy()

# --- Suavizado de la curva ---
# Convertimos fechas a números para la interpolación
data_plot['Fecha_num'] = pd.to_datetime(data_plot['Fecha']).map(pd.Timestamp.toordinal)
x_smooth, y_smooth = suavizar_curva(data_plot['Fecha_num'], data_plot['tasa'])
x_fechas = [pd.Timestamp.fromordinal(int(d)) for d in x_smooth] # Convertir de nuevo a fechas

# --- Gráfica con estilo personalizado ---
plt.figure(figsize=(12, 6), facecolor='#f0f4f8') # Color de fondo de la app
ax = plt.gca()
ax.set_facecolor('#ffffff') # Fondo blanco del gráfico

# 1. Bandas horizontales de riesgo (Gradiente de color de fondo)
# Los límites (y1, y2) deben ajustarse a tus datos reales de riesgo
plt.axhspan(0, 2.5, facecolor='#e0f7fa', alpha=0.4, label='Baja')       # Zona Baja (Azul claro)
plt.axhspan(2.5, 4.5, facecolor='#fff9c4', alpha=0.4, label='Media')    # Zona Media (Amarillo claro)
plt.axhspan(4.5, data_plot['tasa'].max() + 2, facecolor='#ffecb3', alpha=0.4, label='Alta') # Zona Alta (Naranja claro)

# 2. Línea principal y área sombreada
color_linea = '#26a69a' # Color turquesa de tu diseño
plt.plot(x_fechas, y_smooth, color=color_linea, linewidth=3)
plt.fill_between(x_fechas, y_smooth, 0, color=color_linea, alpha=0.2) # Sombreado bajo la curva

# 3. Personalización del gráfico
plt.title(f"Tasa de Fraude - Últimos {periodo_dias} Días", fontsize=16, fontweight='bold', loc='left', color='#37474f', pad=20)
plt.ylabel("Tasa (%)", color='#546e7a')
plt.grid(axis='y', linestyle='--', color='#cfd8dc')
plt.xticks(rotation=0, color='#546e7a')
plt.yticks(color='#546e7a')

# Eliminar bordes innecesarios para un look más limpio
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

## Segmentación de Transacciones/Clientes (Clustering K-Means)

### Subtask:
Determinar el número óptimo de clusters (K) para el algoritmo K-Means usando el método del codo.


**Reasoning**:
To determine the optimal number of clusters using the elbow method, I will import KMeans, iterate through a range of k values, fit the KMeans model to the preprocessed data (`df_fase4`), and collect the inertia for each k. This will allow me to plot the inertia against k and identify the 'elbow' point.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Inicializa una lista vacía para almacenar los valores de inercia
inertia_values = []

# 2. Itera un bucle para valores de k desde 1 hasta 10
# Se recomienda que el rango de k no sea muy grande para el método del codo
for k in range(1, 11):
    # a. Crea una instancia de KMeans con el valor actual de k y random_state
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    # b. Ajusta el modelo KMeans a df_fase4
    kmeans.fit(df_fase4)
    # c. Agrega la inercia del modelo a la lista
    inertia_values.append(kmeans.inertia_)

# 3. Crea un gráfico de línea para visualizar el método del codo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia_values, marker='o', linestyle='-', color='#4a90e2')
plt.title('Método del Codo para Determinar el Número Óptimo de Clusters (K)', fontsize=16, fontweight='bold')
plt.xlabel('Número de Clusters (K)', fontsize=12)
plt.ylabel('Inercia', fontsize=12)
plt.xticks(range(1, 11))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## Segmentación de Transacciones/Clientes (Clustering K-Means)

### Subtask:
Aplicar el algoritmo K-Means con el número óptimo de clusters (k=4) y visualizar los resultados mediante un gráfico de dispersión, utilizando PCA para reducir la dimensionalidad. Finalmente, proporcionar tablas resumen de las características de cada cluster.


**Reasoning**:
To apply K-Means clustering, I will first import the necessary libraries, then instantiate and fit the KMeans model with k=4 to `df_fase4`, add the cluster labels to the DataFrame, and finally perform PCA for dimensionality reduction to prepare for visualization.



```python
from sklearn.decomposition import PCA
import seaborn as sns

# 1. Aplicar K-Means con k=4
kmeans_model = KMeans(n_clusters=4, random_state=42, n_init='auto')
df_fase4['Cluster'] = kmeans_model.fit_predict(df_fase4)

print("✅ K-Means aplicado. Clusters asignados a df_fase4.")

# 2. Reducción de dimensionalidad con PCA
pca = PCA(n_components=2, random_state=42)
df_pca = pd.DataFrame(pca.fit_transform(df_fase4.drop(columns=['Cluster'])), columns=['Componente Principal 1', 'Componente Principal 2'])

# Añadir la columna de cluster a df_pca para la visualización
df_pca['Cluster'] = df_fase4['Cluster'].values

print("✅ PCA realizado y df_pca creado.")
```

**Reasoning**:
To visualize the clusters, I will create a scatter plot of the two principal components, coloring the points by their assigned cluster. Then, I will generate a summary table by calculating the mean of the features for each cluster to understand their characteristics.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 3. Visualizar los clusters resultantes
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='Componente Principal 1',
    y='Componente Principal 2',
    hue='Cluster',
    data=df_pca,
    palette='viridis',
    s=50,
    alpha=0.7
)
plt.title('Clusters de Transacciones (PCA Reducido)', fontsize=16, fontweight='bold')
plt.xlabel('Componente Principal 1', fontsize=12)
plt.ylabel('Componente Principal 2', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Cluster')
plt.show()

print("✅ Gráfico de dispersión de clusters generado.")

# 4. Proporcionar tablas resumen de las características de cada cluster
# Eliminar la columna 'Cluster' de la copia temporal para calcular las medias de las características originales
# Asegurarse de que df_fase4 tiene la columna 'Cluster' antes de agrupar
if 'Cluster' in df_fase4.columns:
    resumen_clusters = df_fase4.groupby('Cluster').mean()
    print("\n--- Resumen de Características por Cluster ---")
    display(resumen_clusters)
else:
    print("Error: La columna 'Cluster' no se encontró en df_fase4.")

print("✅ Tablas resumen de características por cluster generadas.")
```

## Análisis de Correlación entre Variables Numéricas

### Subtask:
Crear un mapa de calor (heatmap) para visualizar la matriz de correlación entre las variables numéricas más relevantes (`Monto`, `Saldo_Cuenta`, `Edad_Cliente`, `Ingresos`).


**Reasoning**:
To visualize the correlation matrix, I need to select the specified numerical columns from `df_fase3_completo`, calculate their correlation, and then plot it using a heatmap with annotations and a title.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Selecciona las columnas numéricas relevantes
columnas_numericas_corr = ['Monto', 'Saldo_Cuenta', 'Edad_Cliente', 'Ingresos']
df_corr = df_fase3_completo[columnas_numericas_corr]

# 2. Calcula la matriz de correlación
matriz_correlacion = df_corr.corr()

# 3. Crea un mapa de calor (heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# 4. Asegúrate de incluir anotaciones y un título descriptivo
plt.title('Matriz de Correlación de Variables Numéricas Relevantes', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```
